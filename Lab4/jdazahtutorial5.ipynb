```python
por_mes = tesla.resample('M').mean()
print(por_mes)
por_dia = tesla.resample('D').std()
print(por_dia)
por_semana = tesla.resample('W').mean()
print(por_semana)
```

```python
cuatro_dias = tesla.resample('4D').mean()
tres_semanas  = tesla.resample('3W').mean()
dos_meses = tesla.resample('2M').mean()

print(cuatro_dias)
print(tres_semanas)
print(dos_meses)
```

```python
std = tesla.resample('W').std()    
max = tesla.resample('D').max()    
min = tesla.resample('M').min()    

print(std)
print(max)
print(min)
```

calcular el retorno semanal del stock

```python
ultimo_dia = tesla.resample('W').agg(lambda x : x[-1])
print(ultimo_dia)
```

para calcular las tasas de retorno (en este caso semanales)

```python
semana_retorno = tesla.resample('W').agg(lambda x : x[-1]/x[1]-1)
print(semana_retorno)
```

Calculos rapidos...


```python
print(semana_retorno.mean())
print(semana_retorno.std())
print(semana_retorno.max())
```

otros dos metodos son `diff` y `pcd_change`,con los cuales se puede calcular la diferencia entre elementos consecutivos y el porcentaje de cambio

```python
print(ultimo_dia.diff())
print(ultimo_dia.pct_change())
```

Para quitar los nan usamos la funcion `fillNan`

```python
retorno_diario = ultimo_dia.pct_change()
print(retorno_diario.fillna(0))
```

otra forma es llenarla con `bfill`

```python
retorno_diario = ultimo_dia.pct_change()
print(retorno_diario.fillna(method='bfill'))
```

O simplemente los quitamos de la serie con `dropna()`

```python
retorno_diario = ultimo_dia.pct_change().dropna()
print(retorno_diario)
```

#Dataframes

```python
import pandas as pd
import pandas_datareader as dr
import numpy as np
```

```python
diccionario = {'TSLA': [143.5,  144.09, 142.73, 144.18, 143.77],
        'GOOG': [898.7,  911.71, 906.69, 918.59, 926.99],
        'MSFT':  [155.58, 153.67, 152.36, 152.94, 153.49]}
fechas = pd.date_range('2018-07-03', periods = 5, freq = 'W')
df = pd.DataFrame(diccionario, index = fechas)
print(df,type(df))
```

```python
df = dr.get_data_yahoo('TSLA')
print(df.Close.tail(15))
print(df['Adj Close'].tail(10))
```

```python
tesla_2020 = df['2020']
tesla_semana = tesla_2020.resample('W').agg(lambda x: x[-1])
print(tesla_semana)
```

```python
tesla_bar = tesla_semana[['Open', 'Volume', 'Low', 'Close']]
print(tesla_bar)
```

```python
print(tesla_semana.loc['2020-01':'2021-01', ['Open', 'High']])
```

```python
above = tesla_bar[tesla_bar.Close > np.mean(tesla_bar.Close)]
print(above)
```

```python
tesla_bar['rate_return'] = tesla_bar.Open.pct_change()
print(tesla_bar)
```

```python
perdido = tesla_bar.isnull()
print(perdido)
print('---------------------------------------------')
print(perdido.describe())
```

```python
print(perdido[perdido.rate_return == True])
print(perdido[perdido.rate_return == False])
```

```python
quitar = tesla_bar.dropna()
print(quitar)
print('\n--------------------------------------------------\n')
fill = tesla_bar.fillna(0)
print(fill)
```

el parametro Axis unira los dos dataframes por columnas

```python
s1 = pd.Series([163.55, 144.09, 645.73, 144.18, 163.57], name = 'MCST')
s2 = pd.Series([898.7, 615.71, 906.69, 968.559, 926.99], name = 'IBM')
data_frame = pd.concat([s1, s2], axis = 1)
print(data_frame)
```

```python
log_price = np.log(tesla_bar.Close)
log_price.name = 'log_price'
print(log_price)
print('\n--------------------------------------------\n')
concatenacion = pd.concat([tesla_bar, log_price], axis = 1)
print(concatenacion)
```

```python
df_volume = tesla_bar.loc['2020-01':'2020-04', ['Volume', 'rate_return']].resample('M').agg(lambda x: x[-1])
print(df_volume)
print('\n-------------------------------------------\n')
df_2020 = tesla_bar.loc['2020-05':'2017-09', ['Open', 'Low', 'Close']].resample('M').agg(lambda x: x[-1])
print(df_2020)
# tesla_bar 
```

```python
concatenar = pd.concat([tesla_bar, df_volume], axis = 1)
print(concatenar)
```

```python
concatenar = pd.concat([tesla_bar, df_volume], axis = 1, join = 'inner')
print(concatenar)
```

Append concatena dos dataframes por axis = 0

```python
append = tesla_bar.append(df_2020)
print(append)
```

si en algun caso cuando se concantena o se unen dos dataframes, hay columnas repetidas, lo que hacemos es renombrarlas

```python
df_2020.columns = ['Open', 'Low', 'Close']
concatenar = pd.concat([tesla_bar, df_2020], axis = 0)
print(concatenar)
```
